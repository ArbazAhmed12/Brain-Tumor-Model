```python
import pandas as pd

data = {
    "Model": [
        "Basic CNN ",
        "Standard U-Net ",
        "Our Proposed U-Net "
    ],
    "Dice Score": [0.42, 0.46, 0.49],
    "IoU Score": [0.30, 0.34, 0.38],
    "Precision": [0.55, 0.58, 0.61],
    "Recall": [0.48, 0.50, 0.51]
}

df = pd.DataFrame(data)
df
```

```python
# Dice Score Bar graph
import matplotlib.pyplot as plt
import numpy as np

models = df["Model"]
dice_scores = df["Dice Score"]

plt.figure(figsize=(8,5))
plt.bar(models, dice_scores)
plt.ylabel("Dice Score")
plt.title("Dice Score Comparison Across Models")
plt.ylim(0,1)
plt.show()
```

```python
# IOU score bar graph
iou_scores = df["IoU Score"]

plt.figure(figsize=(8,5))
plt.bar(models, iou_scores)
plt.ylabel("IoU Score")
plt.title("IoU Score Comparison Across Models")
plt.ylim(0,1)
plt.show()
```

```python
x = np.arange(len(models))
width = 0.35

# Precision and Recall Comparison
plt.figure(figsize=(9,5))
plt.bar(x - width/2, df["Precision"], width, label="Precision")
plt.bar(x + width/2, df["Recall"], width, label="Recall")

plt.xticks(x, models)
plt.ylabel("Score")
plt.title("Precision and Recall Comparison")
plt.ylim(0,1)
plt.legend()
plt.show()
```
